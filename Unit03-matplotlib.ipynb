```python
import matplotlib.pyplot as plt
plt.title('legend')
plt.plot([10, 20, 30, 40], label='asc')    #   asc 
plt.plot([40, 30, 20, 10], label='desc')   #   desc 
plt.legend(loc = 4)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('color')  #  
#  
plt.plot([10, 20, 30, 40], color ='r', label='skyblue')
plt.plot([40, 30, 20, 10], 'pink', label='pink')
plt.legend()   #  
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('linestyle')  #  
#  dashed 
plt.plot([10, 20, 30, 40], color ='r', linestyle='--', label='dashed')
#  dotted 
plt.plot([40, 30, 20, 10], color ='g', ls=':', label='dotted')
plt.legend()  #  
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('marker')  #  
plt.plot([10, 20, 30, 40], 'r.', label='circle')  #    
#    
plt.plot([40, 30, 20, 10], 'g^', label='triangle up')
plt.legend()  #  
plt.show()
```
